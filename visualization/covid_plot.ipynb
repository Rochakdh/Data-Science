```python
for index,row in df.drop_duplicates(subset=['location'],keep='last').iterrows():
    if row['location'] in saarc_country:
        print(row[['location','total_deaths']])
```

```python
quantity=[]
label=[]
saarc_country = [ 'Afghanistan', 'Bangladesh', 'Maldives', 'Pakistan','India','Sri Lanka','Nepal']
for index,row in df.drop_duplicates(subset=['location'],keep='last').iterrows():
    if row['location'] in saarc_country:
        quantity.append(row[['total_deaths']][0])
        label.append(row[['iso_code']][0])
print(label)
print(quantity)
plt.pie(quantity,labels=label ,autopct='%1.1f%%',radius = 3,shadow=True,textprops={'fontsize': 10},colors=['purple','yellow','orange','teal','green','brown'])
plt.show()
plt.bar(label,quantity,width=0.3)
plt.show()
```

```python
newcases=[]
for index,row in df.iterrows():
    if row['location']=='India':
        newcases.append(row['new_cases'])
# print(newcases)
plt.plot(newcases,color='red',linestyle='solid',linewidth=0.8)
plt.show()
```
